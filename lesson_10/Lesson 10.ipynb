### Яка різниця між list, set та tuple?



Що та коли використовувати:

- Список:
    - Послідовність буде змінюватися.
    - Планується поступово додавати нові елементи до послідовності або видаляти старі.
- Множини:
    - Послідовність не планується змінювати.
    - Все, що потрібно від послідовності, це можливість по черзі перебирати постійний набір елементів.
    - Потрібна послідовність елементів для її призначення як ключ словника. Оскільки списки — це тип даних, що змінюється, їх не рекомендовано застосовувати як ключі словника.
    - Важлива швидкість виконання операцій з послідовністю: через відсутність можливості зміни, кортежі працюють набагато швидше за списки.
