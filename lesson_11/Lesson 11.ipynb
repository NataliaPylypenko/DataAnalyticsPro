```python
username = input("Введіть ім'я користувача: ")
password = input("Введіть пароль: ")

password_correct = False

while not password_correct:
    if len(password) < 8:
        print("Пароль надто короткий\n")
        password = input("Введіть пароль ще раз: ")
    elif username in password:
        print("Пароль містить ім'я користувача\n")
        password = input("Введіть пароль ще раз: ")
    else:
        print(f"Пароль для користувача {username} встановлено")
        password_correct = True
```

```python
#using break with a password check loop
username = input("Введіть ім'я користувача: ")
password = input("Введіть пароль: ")

while True:
    if len(password) < 8:
        print("Пароль надто короткий\n")
    elif username in password:
        print("Пароль містить ім'я користувача\n")
    else:
        print("Пароль для користувача {} встановлено".format(username))
        # завершує цикл while
        break
    password = input("Введіть пароль ще раз: ")
```

### try/except

```python
print(2/0)
```

```python
print('test' + 2)
```

```python
try:
    2/0
except ZeroDivisionError:
    print('You cant divide by zero')
```

```python
#приклад з декількома exception handlers
num1 = input("Введіть перше число: ")
num2 = input("Введіть друге число: ")

try:
    print("Результат: ", int(num1) / int(num2))
except ValueError:
    print("Вводьте лише числа")
except ZeroDivisionError:
    print("На нуль ділити не можна")
```

```python
#можемо обєднати вийнятки
num1 = input("Введіть перше число: ")
num2 = input("Введіть друге число: ")

try:
    print("Результат: ", int(num1)/int(num2))
except (ValueError, ZeroDivisionError):
    print("Щось пішло не так...")
```

### try/except/else

```python
num1 = input("Введіть перше число: ")
num2 = input("Введіть друге число: ")
try:
    result = int(num1)/int(num2)
except (ValueError, ZeroDivisionError):
    print("Щось пішло не так...")
else:
    print("Результат: ", result)
```

### try/except/finally

```python
num1 = input("Введіть перше число: ")
num2 = input("Введіть друге число: ")
try:
    result = int(num1)/int(num2)
except (ValueError, ZeroDivisionError):
    print("Щось пішло не так...")
else:
    print("Результат: ", result**2)
finally:
    print("The End")
```

## Python Comprehensions

```python
#list comprehensions
items = ["10", "20", "30", "1", "11", "100"]

#loop for
vlans = []
for vl in items:
    vlans.append(int(vl))

#same but using comprehensions
vlans = [int(vl) for vl in items]
```

Різниця між циклом та генератором:



```python
# можна використовувати умови
items = ['10', '20', '30', '1', '11', 'ab', '100']

vlans = [int(vl) for vl in items if vl.isdigit()]
print(vlans)
```

Це буде теж саме що й наступне:



```python
#dictionary comprehension

#for loop
d = {}
for num in range(1, 11):
    d[num] = num**2

print(d)

#comprehension
#d = {num: num**2 for num in range(1, 11)}
```

```python
#робило маленькими літерами ключі

#option 1
r1 = {
    'IOS': '15.4',
    'IP': '10.255.0.1',
    'hostname': 'london_r1',
    'location': '21 New Globe Walk',
    'model': '4451',
    'Vendor': 'Cisco'
}

lower_r1 = {}

for key, value in r1.items():
    lower_r1[key.lower()] = value
```

```python
#option 2 - comprehensions
lower_r1 = {key.lower(): value for key, value in r1.items()}
```

```python
#вкладений вираз
london_co = {
    "r1": {
        "hostname": "london_r1",
        "IOS": "15.4",
        "IP": "10.255.0.1",
    },
    "r2": {
        "hostname": "london_r2",
        "IOS": "15.4",
        "IP": "10.255.0.2",
    },
    "sw1": {
        "hostname": "london_sw1",
        "IOS": "3.6.XE",
        "IP": "10.255.0.101",
    },
}

lower_london_co = {}

for device, params in london_co.items():
    lower_london_co[device] = {}
    for key, value in params.items():
        lower_london_co[device][key.lower()] = value
```

```python
#те саме але красивіше
result = {device: {key.lower(): value for key, value in params.items()}
          for device, params in london_co.items()}
```
