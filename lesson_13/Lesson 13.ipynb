## Що таке ООП?

**Об’єктно-орієнтоване програмування (ООП)** — це парадигма програмування, зосереджена навколо концепції об’єктів, які можуть містити як дані (атрибути), так і методи (функції). Python активно використовує принципи ООП.


Why **OOP** is Useful:

* **Модульність**: Ви можете розбивати програми на менші керовані частини (об’єкти).
* **Повторне використання**: класи та об’єкти можна повторно використовувати в різних частинах програми або інших програмах.
* **Гнучкість**: ООП сприяє коду, який легше змінювати та розширювати.
* **Обслуговування**: ООП полегшує налагодження та підтримку коду, особливо у великих програмах.

У Python **усе є об’єктом**, оскільки Python дотримується парадигми об’єктно-орієнтованого програмування.
Причини твердження:

* **Unified Design**:
    * Python обробляє все — від простих типів даних, таких як цілі чи рядки, до складніших конструкцій, таких як функції та класи — як об’єкти. Це створює послідовну уніфіковану модель для обробки даних і поведінки.

* **Attributes and Methods**:
    * Кожен об’єкт у Python має пов’язані з ним атрибути (дані) і методи (функції). Наприклад, рядок має такі методи, як .upper() або .split(), і навіть цілі числа мають власні методи. Ви можете отримати доступ до цих методів за допомогою крапкової нотації.

* **Inheritance and Reusability**:
    * Оскільки все є об’єктом, ядро ​​функціональності Python покладається на успадкування класів. Наприклад, цілі числа, числа з плаваючою точкою та списки є екземплярами відповідних класів. Це дозволяє Python повторно використовувати код і пропонувати узгоджену функціональність.

* **Dynamic Typing**:
    * Динамічна типізація Python побудована на цій об’єктній структурі. Вам не потрібно оголошувати типи змінних, оскільки інтерпретатор розуміє, що всі змінні є об’єктами, і може автоматично керувати їхнім типом.

* **Real-Life Example**:
    * Подумайте про клас як про форму для печива, а печиво – про об’єкти. Формочка (клас) визначає форму (атрибути та методи), але печеньки (об’єкти) є окремими екземплярами цієї форми.

### Типи конструкторів

- **Конструктор за замовчуванням** — це простий конструктор, який не приймає жодних параметрів. Його визначення має тільки один параметр, яким є посилання на створюваний об’єкт.
- **Параметризований конструктор** — це конструктор з параметрами: перший параметр — це посилання на створюваний об’єкт (self), а інші параметри надаються програмістом.

```python
class Pet1:
#	by default
	def __init__(self):
		self.name = 'Tuzik'

class Pet2:
#	parametric
	def __init__(self, name, pet_type, age):
		self.name = name
		self.pet_type = pet_type
		self.age = age
```

**Переваги** використання конструкторів:
- **Ініціалізація об’єктів**. Конструктори використовуються для ініціалізації об’єктів класу. Вони дозволяють встановити значення за замовчуванням для членів класу, а також ініціалізувати об’єкт користувацькими даними.
- **Простота реалізації**. Конструктори легко реалізувати в Python за допомогою методу __init__().
- **Покращення читабельності коду**. Конструктори покращують читабельність коду, оскільки дають зрозуміти, які значення ініціалізуються і як вони ініціалізуються.
- **Інкапсуляція**. Конструктори можуть використовуватися для забезпечення інкапсуляції, гарантуючи, що члени об’єкта ініціалізуються коректно та контрольовано.

**Недоліки** використання конструкторів:
- **Обмежена функціональність**. Конструктори в Python мають обмежену функціональність порівняно з конструкторами з інших мов програмування. Наприклад, у Python немає конструкторів з модифікаторами доступу, такими як public, private або protected.
- **Конструктори можуть бути непотрібними**. У деяких випадках конструктори можуть бути непотрібними, оскільки є достатньо значень за замовчуванням для членів класу. У таких випадках використання конструктора може призвести до зайвого ускладнення коду.

В цілому, конструктори в Python можуть бути корисними для ініціалізації об’єктів та забезпечення інкапсуляції. Однак вони не завжди є необхідними та мають обмежену функціональність у порівнянні з конструкторами в інших мовах програмування.

## 1. Encapsulation

**Інкапсуляція** — це об’єднання даних (атрибутів) і методів (функцій), які працюють з даними, в один блок, тобто клас.

Why it’s important:
* Захист даних: запобігає ненавмисному або шкідливому доступу або зміні даних. Дані можуть бути змінені лише через чітко визначені інтерфейси (методи getter та setter).
* Спрощує інтерфейси: об’єкти забезпечують чистий і зрозумілий інтерфейс для взаємодії з ними, не турбуючись про те, як вони працюють всередині.

### Приховування атрибутів

В багатьох мовах програмування, які підтримують парадигму ООП, існують спеціальні модифікатори доступу атрибутів. Вони явно вказують, чи можна мати доступ до певного атрибуту класа "ззовні", або ж цей атрибут доступний тільки всередині класа.

В Python механізму модифікаторів доступу **не існує**. Для імітації приховування атрибутів в Python використовується **домовленість** згідно якої якщо ідентифікатор атрибута починається з знака підкреслення, то цей атрибут призначено виключно для внутрішнього використання.

**ВАРТО ПАМЯТАТИ**: **домовленість** — це не синтаксичне правило мови програмування, і при великому бажанні її можна порушити

```python
class Person:
    def __init__(self, age):
        self._age = age

p = Person(35)
p._age
```

Для того щоб поменше порушувати домовленості, можна ще більше приховати атрибут класа. Для цього його ідентифікатор має починатись не з одного, а з двох знаків підкреслення.

```python
class Person:
    def __init__(self, age):
        self.__age = age

p = Person(35)
p.__age
```

### Сетери і гетери

По факту, це ***методи**, які використовуються для **доступу та оновлення** приватних атрибутів (тих, що мають префікс _ або __) класу. Хоча Python не вимагає використання геттерів і сеттерів, як деякі інші мови програмування (наприклад, Java), вони надають корисний механізм для контролю доступу до атрибутів і модифікації.

В об'єктно-орієнтованому програмуванні прийнято імена методів для вилучення даних починати зі слова get ("взяти"), а імена методів, в яких полям присвоюються певні значення — зі слова set ("встановити"). А самі методи часто називають відповідно сетерами і гетерами. Існують також делетери — методи для видалення (delete) полів класа.

```python
class Person:
    def __init__(self, age):
        self.set_age(age)

    #getter
    def get_age(self):
        return self.__age
    
    #setter
    def set_age(self, age):
        if age > 0:
            self.__age = age

p = Person(35)
age = p.get_age()
age
```

```python
p.set_age(-35)
p.get_age()
```

## 2. Abstraction
**Абстракція** зосереджується на приховуванні складних деталей реалізації та показі лише основних характеристик об’єкта.

В ООП класи та об'єкти є основними формами абстракції. Клас визначає загальні властивості та методи, які можуть бути використані об'єктами цього класу, а об'єкт є конкретною реалізацією класу та містить властивості та методи, визначені у класі. Вона дозволяє створювати об'єкти, які можна використовувати у різних контекстах, не знаючи всіх деталей реалізації.

Ідея:
* Зменшує складність: користувачам класу чи об’єкта не потрібно знати, як він працює всередині.
* Покращує фокус: допомагає програмістам зосередитися на тому, «що» об’єкт робить, а не на тому, «як» він це робить.

    


```python
#abstract class
class Animal:
    name = ''
    age = ''

class Dog(Animal):  # Dog class inherits from Animal
    def get_info(self):
        print(self.name, self.age)

class Cat(Animal):  # Cat class inherits from Animal
    def get_info(self):
        print(self.name, self.age)

class Cow(Animal):  # Cat class inherits from Animal
    def get_info(self):
        print(self.name, self.age)
```

## 3. Inheritance
**Спадкування** дозволяє одному класу (дочірньому/підкласу) успадковувати атрибути та методи іншого класу (батьківського/суперкласу).



Яка **користь** від спадкування в Python?
- Оскільки дочірній клас може успадковувати функціонал батьківського класу, це дозволяє повторно використовувати код без його дублювання.
- Розробивши функціонал, ми можемо просто його успадкувати. Не потрібно наново винаходити велосипед. Це робить код чистішим і простішим у підтримці.
- Оскільки в дочірній клас можна додавати власний функціонал, то можна успадковувати лише потрібний функціонал батьківського класу, а все інше дописати в конкретному (дочірньому) класі.
- Забезпечує природні ієрархічні зв’язки між об’єктами, що відображають зв’язки в реальному світі.


```python
#superclass
class Animal:
    name = ''
    def eat(self):
        print("East")

#subclass
class Dog(Animal):  # Dog class inherits from Animal
    def sound(self):
        print("Barks!")

class Cat(Animal):  # Cat class inherits from Animal
    def sound(self):
        print("Meaws!")

class Cow(Animal):  # Cat class inherits from Animal
    def sound(self):
        print("Mooos!")
```

З попереднього прикладу ми бачимо, що об’єкт підкласу може отримати доступ до методу суперкласу.

**Але якщо один і той самий метод присутній і в суперкласі, і в підкласі?**

У цьому випадку метод у підкласі перевизначає метод у суперкласі. Ця концепція відома як **перевизначення методів** у Python.

```python
#example
```

### Метод super() в спадкуванні

Раніше ми бачили, що метод з одним і тим самим ім’ям у підкласі перевизначає метод у суперкласі. Однак, якщо нам потрібно отримати доступ до методу суперкласу підкласу, ми можемо скористатися додатковим методом **super()**.

```python
#example
class Pet:
    # Атрибути та метод батьківського класу
    name = ""
    def eat(self):
        print("I can eat")
 
# Успадковуємо від класу Animal
class Dog(Pet):
    # Перевизначаємо метод eat()
    def eat(self):
        # Викликаємо метод eat() батьківського класу за допомогою додаткового методу super()
        super().eat()
        print("I like to eat bones")
 
# Створюємо об'єкт дочірнього класу
labrador = Dog()
labrador.eat()
```

Тонкощі спадкування:
- **Множинне успадкування** - один клас може успадковуватися відразу від декількох батьківських класів. Це називається множинним успадкуванням.
- **Багаторівневе спадкування** - дочірній клас може успадковувати не лише батьківський клас, а й інший дочірній клас. Така форма спадкування відома як багаторівневе спадкування.
- **Порядок виклику методів (MRO)** - якщо два батьківських класи мають однакове ім’я методу та дочірній клас (який успадковує 2 батьківські класи) викликає цей метод, Python використовує MRO (порядок виклику методів) для виклику потрібного методу.

```python
#множинне спадкування
class Mammal:
    def mammal_info(self):
        print("Mammals can give direct birth.")
 
class WingedAnimal:
    def winged_animal_info(self):
        print("Winged animals can flap.")
 
class Bat(Mammal, WingedAnimal):
    pass
 
# Створюємо об'єкт класу Bat
b1 = Bat()
 
b1.mammal_info()
b1.winged_animal_info()
```

```python
#багаторівневе спадкування
class SuperClass:
    def super_method(self):
        print("Super Class method called")
 
# Визначаємо клас, який успадковує SuperClass
class DerivedClass1(SuperClass):
    def derived1_method(self):
        print("Derived class 1 method called")
 
# Визначаємо клас, який успадковує DerivedClass1
class DerivedClass2(DerivedClass1):
    def derived2_method(self):
        print("Derived class 2 method called")
 
# Створюємо об'єкт класу DerivedClass2
d2 = DerivedClass2()
 
# Викликаємо методи
d2.super_method()  
d2.derived1_method()  
d2.derived2_method()
```

```python
#порядок виконання методів
class SuperClass1:
    def info(self):
        print("Super Class 1 method called")
 
class SuperClass2:
    def info(self):
        print("Super Class 2 method called")
 
class Derived(SuperClass1, SuperClass2):
    pass
 
d1 = Derived()
d1.info()  
```

Тут обидва класи SuperClass1 та SuperClass2 мають метод info(). Коли метод info() викликається за допомогою об’єкта d1 класу Derived, Python використовує MRO, щоб визначити, метод якого класу слід викликати.

В цьому випадку MRO визначає, що методи повинні успадковуватися спочатку від найлівішого батьківського класу. Тому викликається метод info() класу SuperClass1, а не класу SuperClass2

## 4. Polymorphism

**Поліморфізм** дозволяє взаємозамінно використовувати методи між різними об’єктами, навіть якщо вони належать до різних класів. Це означає «багато форм» і стосується здатності використовувати загальний інтерфейс для різних базових форм (типів даних).

Why it’s important:
* Гнучкість: дозволяє одній функції чи методу обробляти різні типи об’єктів.
*  Узагальнення коду: дозволяє писати більш загальний код, який може працювати з об’єктами різних класів, якщо вони мають спільний інтерфейс (наприклад, імена методів).

```python
class Cat:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def make_sound(self):
    print("Meow!")

class Dog:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def make_sound(self):
    print("Bark!")

cat = Cat('Surskiy', 10)
dog = Dog('Tuzik', 8)

for x in (cat, dog):
    print(x.name)
    x.make_sound()
```

Підсумок:

* **Інкапсуляція**: захищає дані, поєднуючи їх із методами.
* **Абстракція**: приховує непотрібні деталі, показуючи лише те, що важливо.
* **Успадкування**: дозволяє новим класам успадковувати функціональність від існуючих класів.
* **Поліморфізм**: дозволяє виконувати ту саму операцію різними способами залежно від об’єкта.

Функції класу, що починаються з подвійного підкреслення __, називаються **спеціальними функціями** в Python. Спеціальні функції визначаються інтерпретатором Python і використовуються для реалізації певних можливостей чи поведінки.

Вони називаються **функціями з подвійним підкресленням**, тому що мають префікс та суфікс у вигляді подвійного підкреслення, наприклад __init__() або __add__().

Ось деякі зі спеціальних функцій, які доступні в Python:



Можна змінити спосіб роботи операторів з користувацькими типами даних. Наприклад, ми можемо зробити так, щоб оператор + виконував арифметичне додавання двох чисел або об’єднання двох списків або конкатенацію двох рядків.

Ця можливість в Python, що дозволяє одному і тому ж оператору виконувати різні дії в залежності від контексту, називається **перевантаженням оператора**.
приклад перевантаження оператора + в Python. Для цього нам необхідно визначити у класі функцію __add__().

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
 
    def __str__(self):
        return "({0},{1})".format(self.x, self.y)
 
    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x, y)
 
p1 = Point(1, 2)
p2 = Point(2, 3)
 
print(p1 + p2)
```

Коли ми виконуємо p1 + p2, Python викликає метод p1.__add__(p2), який, в свою чергу, є Point.__add__(p1,p2). Після цього операція додавання виконується по кожному параметру окремо.

## Декоратори

У Python **декоратори** є потужним механізмом для зміни поведінки функцій або методів, не змінюючи їх вихідний код. Декоратори часто використовуються для додавання функціональності до існуючих функцій, логування, перевірок прав доступу або перетворення результатів.

```python
#є функція подарунку
def gift_function():
    print('Я — подарунок!')

gift_function()
```

```python
#є функція яка є святковою обгорткою нашого подарунку

def wrap_function(gift_to_wrap_function):
    print('Я — святкова обгортка! Я обгорну подарунок.')
    gift_to_wrap_function()
    print('Подарунок обгорнуто!')

wrap_function(gift_function)
```

```python
#уявлюємо людину, яка займається обгорткою подарунків. назвемо її декоратор
def decorator_function(gift_to_wrap_function):
    def wrap_function():
        print('Я — святкова обгортка! Я обгорну подарунок.')
        gift_to_wrap_function()
        print('Подарунок обгорнуто!')
    return wrap_function

gift_function  = decorator_function(gift_function)
gift_function()
```

**Ідея**: взяли функцію, за допомогою декоратора додали до неї певний функціонал, і таким чином отримали цю функцію, але вже "покращену".

**Декоратори** — дуже корисна і зручна штука, і на практиці декоратори застосовуються досить часто. Тому для декораторів в Python придумали спеціальний синтаксис, який дозволяє використовувати декоратори більш зручно.

```python
#декоруємо новий подарунок
def gift_iphone():
     print('Я — айфон!')

gift_iphone = decorator_function(gift_iphone)
gift_iphone()
```

```python
#ідентичне рішення за допомогою декоратора
@decorator_function
def gift_iphone():
    print('Я — айфон!')

gift_iphone()
```

Тобто спочатку вказуємо ім'я декоратора після значка @, і на наступному рядку ту функцію, яку треба декорувати вищенаведеним декоратором.

```python
#ланцюжок декораторів - створюємо бутерброд

def bread(func):
    def wrapper():
        print('Хліб')
        func()
        print('Хліб')
    return wrapper

def salad(func):
    def wrapper():
            print('Зеленина')
            func()
            print('Зеленина')
    return wrapper

@bread
@salad
def stake():
    print('Мясо')

stake()
```

## Статичні методи, методи класу та екземпляра класу

Python пропонує три види методів: статичні, класу та екземпляра класу. Про статичні методи та методи класу можна прочитати тут: [Декоратори @staticmethod та @classmethod в Python](https://www.it-notes.wiki/python/decorators-staticmethod-and-classmethod/)

**@staticmethod** та **@classmethod** – це вбудовані декоратори Python, що використовуються для зміни поведінки методів в класі.

**Статичний метод** — це метод, прив’язаний до класу, а не екземпляр класу. Він не може отримати доступ або змінити стан класу або екземпляра (атрибути).

Застосування:
- використовується для службових функцій, які не залежать від специфічних для класу даних або змінних екземпляра.
- статичний метод не містить параметри cls або self (та не має доступ до атрибутів класу або екземпляра)
- для виклику статичного метода не обов’язково створювати екземпляр класу. Виклик статичного метода відбувається напряму через ім’я класу: ClassName.method_name(). Виклик через екземпляр, також буде працювати.

https://www.w3schools.com/python/python_datetime.asp

```python
import datetime

class Event:
    def __init__(self, event_name, event_date):
        self.event_name = event_name
        self.event_date = event_date
        self.current_time = self.get_current_time()

    def get_event_info(self):
       print(f'Event info: {self.event_name}, {self.event_date}, {self.current_time}')

    @staticmethod
    def get_current_time():
      return datetime.datetime.now().strftime("%d.%m.%y")

p = Event('open_session', '2024/02/14')
print(p.event_name, p.event_date, p.current_time)
```

**@classmethod** в Python використовується для визначення методів класу (ці методи мають відношення до класу в цілому, а не його окремого екземпляра).

Методи класу в якості першого аргумента приймають сам клас (умовна назва аргумента – cls), завдяки чому мають доступ до атрибутів класу, і можуть використовуватись для зміни стану класу або створення його екземплярів.

Приклад використання методів класу, в якості альтернативних конструкторів, коли є потреба реалізувати кілька варіантів створення екземпляра класу (в самому Python можна створювати лише один метод __init__ в межах класу)

```python
from faker import Faker

class Person:
    fake = Faker('uk_UA')
    name = 'Тарас Полішко'
    citizenship = 'Україна'

    def __init__(self, name, citizenship):
        self.name = name
        self.citizenship = citizenship

    @classmethod
    def get_default_person(cls):
        return cls(cls.name, cls.citizenship)

    @classmethod
    def get_random_person(cls):
        name = cls.fake.name()
        citizenship = cls.fake.country()
        return cls(name, citizenship)
```

- методи класу – get_default_person(cls) та get_random_person(cls), через аргумент cls отримують доступ до атрибутів красу Person (змінні name, citizenship, fake)
- get_default_person та get_random_person повертають дві різні варіації об’єктів Person
- виклик методу __init__ відбувається за допомогою синтаксису cls(argument_value), в прикладі вище це:
    - cls(cls.name, cls.citizenship)
    - cls(name, citizenship)

```python
default_person = Person.get_default_person()
print(default_person)

random_person = Person.get_random_person()
print(random_person)
```

## Змінні класу та екземпляра класу

[Почитати тут](https://medium.com/@pouyahallaj/class-vs-instance-variables-in-python-5573e71c99b5)

- Змінні класу призначені для атрибутів, значення котрих є спільними для всіх його екземплярів
- Змінні екземпляра призначені для даних, котрі відносяться до конкретного екземпляра класу, та не залежать від інших екземплярів



### Змінні класу (class variable):

* Scope: Ці змінні доступні за допомогою імені класу або будь-якого екземпляра класу.

* Use Case: Iдеально, коли ви хочете поділитися даними між усіма об’єктами, створеними з класу.

```python
class Cat:
    pet_type = 'Very nice pet'  # class default variable

    def __init__(self, name, age):
        self.name = name
        self.age = age

cat1 = Cat('Pushok', 5)
cat2 = Cat('Nichka', 3)

print(cat1.pet_type)
print(cat2.pet_type)
```

### Змінна екземпляру класу (instance variable):
* Визначені всередині конструктора (тобто, метод __init__) і містять дані, специфічні для кожного об’єкта.
* Scope: Ці змінні доступні лише через екземпляр, якому вони належать, а не через сам клас.
* Use Case: Використовуйте змінні екземплярів, якщо ви хочете, щоб кожен об’єкт мав власні дані.

```python
class Cat:
    pet_type = 'Very nice pet'  # class default variable

    def __init__(self, name, age):
        self.name = name # instance variable
        self.age = age # instance variable

cat1 = Cat('Pushok', 5)
cat2 = Cat('Nichka', 3)

print(cat1.pet_type)
print(cat2.pet_type)
```

```python
#counting instances - для того щоб рахувати скільки екземплярів наплодили
class Task:
    task_count = 0

    def __init__(self, description):
        self.description = description
        Task.task_count += 1

task1 = Task("Write article")
task2 = Task("Complete project")

print(Task.task_count)  # Output: 2
```

```python
#shared state - спільний стан класу який змінюється від екземпляру до екземпляру
class BankAccount:
    total_balance = 0

    def __init__(self, balance):
        self.balance = balance
        BankAccount.total_balance += balance

acc1 = BankAccount(1000)
acc2 = BankAccount(500)

print(BankAccount.total_balance)
```

```python
class Author:
    def __init__(self, name, surname, year_of_birth):
        self.name = name
        self.surname = surname
        self.year_of_birth = year_of_birth
        self.books = []  # List to store books written by the author

    def get_full_name(self):
        return f'{self.name} {self.surname}'

    def add_book(self, book):
        self.books.append(book)

    def get_bibliography(self):
        if not self.books:
            print(f"{self.name} has not written any books yet.")
        else:
            print(f"Books by {self.get_full_name()}:")
            for book in self.books:
                print(f'-- {book.get_book_info()}')
```

```python
glukh = Author('Dmytro', 'Glukhovskyi', 1979)
glukh.get_bibliography()
```

```python
# Book Class
class Book:
    def __init__(self, title, genre, year, author, language = 'en'):
        self.title = title
        self.genre = genre
        self.year = year
        self.author = author
        self.language = language

        author.add_book(self)  # Add the book to the author's bibliography

    def get_book_info(self):
        return f'{self.title}, {self.year}'

    def get_author(self):
        return f'{self.author.name} {self.author.surname}'

    def get_details(self):
        print(f"-- Title: {self.title}")
        print(f"-- Genre: {self.genre}")
        print(f"-- Language: {self.language}")
        print(f"-- Publication Year: {self.year}")
        print(f"-- Author:", self.get_author())
        print('-'*50)
```

```python
metro_book = Book('Metro 2033', 'post-apocalypse', 2002, glukh)
metro_book.get_details()

fut = Book('Futu.re', 'post-apocalypse', 2013, glukh)
fut.get_details()
```
