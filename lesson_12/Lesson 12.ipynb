## None vs NaN

* **False**: Представляє логічне значення для "брехні". Може вказувати на відсутність чогось у логічних контекстах, але це все одно якесь значення.
* **Порожні колекції ([], {}, '')**
  * Порожні колекції, як-от порожні списки, словники та рядки, вважаються False у логічних контекстах, але це все одно якесь значення.
* **0**: Числовий нуль також вважається False у логічних контекстах, хоча це і справжнє число.
* **None**: Позначає відсутність значення. Python використовує ключове слово None для визначення нульових об’єктів і змінних.
* **NaN**: Спеціальне значення з плаваючою комою, яке використовується для представлення недійсних або невизначених числових результатів.

```python
x = None
print(x == None)   # True
print(x is None)   # True
print(bool(x)) # False (None is treated as False in boolean contexts)
```

```python
from math import nan, isnan

y_1 = float('nan')
y_2 = nan
print(f'y_1 = {y_1}; y_2 = {y_2}')
print('y_1 == y_1 :', y_1 == y_1)      # False (NaN is not equal to anything, not even itself)
print('y_1 == y_2 :', y_1 == y_2)      # False (NaN is not equal to anything, not even itself)
print('isnan(y_1) :', isnan(y_1))      # True
print('bool(y_1)) :', bool(y_1))       # True (NaN is treated as True in boolean contexts)
```

# Functions

## Що таке функція?
* Функція — це блок багаторазового коду, який виконує певне завдання.
* Функції допомагають упорядкувати код, покращити читабельність і зробити його модульним, дозволяючи повторне використання коду без його повторення.

## Навіщо нам функції?
* **Повторне використання коду**: ви можете викликати ту саму функцію в різних частинах коду, не переписуючи його.
* **Модульність**: функції дозволяють розбивати складні проблеми на менші, керовані завдання.
* **Покращена читабельність**: функції роблять код чистішим і легшим для розуміння.
* **Тестування**: Ви можете тестувати функції окремо, не тестуючи всю програму.

## Назви функцій
* Використовуйте малі літери зі словами, розділеними підкресленням (snake_case).
* Назва функції має бути описовою та відображати дію, яку виконує функція.
* Уникайте використання зарезервованих ключових слів Python як імен функцій.

### Що таке map(), zip(), filter()?

Стаття: https://dev.to/codespent/understanding-map-filter-and-zip-in-python-3ifn

#### map():

* Призначення: Застосовує задану функцію до всіх елементів ітератора (наприклад, списку) і повертає об’єкт карти (який є ітератором).
* Syntax: map(function, iterable)
* Випадок використання: коли вам потрібно перетворити кожен елемент у ітерований.


```python
def get_square(a):
    return a**2

numbers = [1, 2, 3]
squares = list(map(get_square, numbers))
print(squares)
```

#### zip():

* Призначення: об’єднує елементи з двох або більше ітерованих (наприклад, списків або кортежів) у кортежі, поєднуючи елементи з однаковим індексом.
* Синтаксис: zip(iterable1, iterable2, ...)
* Випадок використання: коли вам потрібно поєднати елементи з кількох ітерацій.

```python
even_numbers = [2,4]
even_numbers_squared = [4, 8]
zipped_result = zip(even_numbers, even_numbers_squared)
```

#### filter():

* Призначення: фільтрує елементи в ітерації, застосовуючи функцію, яка повертає True або False, і повертає лише ті елементи, для яких функція повертає True.
* Синтаксис: фільтр (функція, ітерація)
* Випадок використання: коли потрібно відфільтрувати елементи на основі умови.

```python
numbers = [1, -2, 3, 4, 5]
even_numbers = list(filter(lambda x: (x % 2 == 0), numbers))
print(even_numbers)
```

## Ще деякі особливості функцій

### *args & **kwargs

Є ще додаткова можливість функцій приймати наступні параметри:
- *args - це скорочення від "arguments" (аргументи). Приймає будь-яку кількість позиційних аргументів, створюючи з них список.
- **kwargs - скорочення від "keyword arguments" (іменовані аргументи). Приймає будь-яку кількість іменованих аргументів, створюючи з них словник.

Кожна з цих конструкцій використовується для розпакування аргументів відповідного типу, що дозволяє викликати функції зі списком аргументів змінної довжини.
Наприклад - створимо функцію, яка вміє виводити результати, набрані учнем у тесті

```python
#ex1
def get_scores(student, *scores):
    print(f"Student Name: {student}")
    for score in scores:
        print(score)

get_scores('Jonathan', 100, 95, 88, 92, 99)
```

```python
#ex2
def sum_of_numbers(*args):
    return sum(args)

print(sum_of_numbers(1, 2, 3, 4, 5, 1, 1, 1, 4, 6))
```

```python
#ex3
def get_pets_name(owner, **pets):
   print(f"Owner Name: {owner}")
   for pet, name in pets.items():
      print(f"{pet}: {name}")

get_pets_name(
   'Jonathan',
   dog = 'Brock', 
   fish = ['Larry', 'Curly'],
   turtle = 'Shelldon'
)
```

```python
#комбінація

def show_info(*args, **kwargs):
    print('Positional arguments:', args)
    print('Keyword arguments:', kwargs)

show_info(1, 2, 3, name = "Alice", age = 25)
```

## Поширені проблеми/нюанси роботи з функціями

```python
#bad example
result = []

def func(items):
    for i in items:
        result.append(i*100)
    return result
```

```python
#good example
def func(items):
    result = []
    for i in items:
        result.append(i*100)
    return result
```

```python
#bad example

from datetime import datetime
import time

def print_current_datetime(ptime=datetime.now()):
    print(f">>> {ptime}")


for i in range(3):
    print("Імітуємо довге виконання...")
    time.sleep(1)
    print_current_datetime()
```

```python
#good example
def print_current_datetime():
    print(f">>> {datetime.now()}")
```

```python
#bad example
def add_item(item, data=[]):
    data.append(item)
    return data

add_item(1)
add_item(2)
add_item(4)
```

```python
#good example
def add_item(item, data = None):
    if data is None:
        data = []
    data.append(item)
    return data

add_item(1)
add_item(2)
```
